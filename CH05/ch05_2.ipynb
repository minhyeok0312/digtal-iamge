```python
from random import randint
```

```python
# 10% noise
noiseRatio = 20
whiteMax = noiseRatio/2
blackMax = noiseRatio
```

```python
noiseimg = gimg.copy()

for h in range(0, gimg.shape[0]):
  for w in range(0, gimg.shape[1]):
    val = randint(0, 100)
    if(val < whiteMax):
      noiseimg[h, w] = 255
    elif(val < blackMax):
      noiseimg[h, w]= 0

cv2_imshow(noiseimg)
```

```python
outimg = cv2.medianBlur(noiseimg, 3)
cv2_imshow(outimg)
```

```python
outimg2 = cv2.medianBlur(noiseimg, 5)
cv2_imshow(outimg2)
```
