```python
img = cv2.imread('/content/gdrive/My Drive/ColabIP/images/Copy of shrinking.bmp')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)

gimg = cv2.cvtColor(RGB_img, cv2.COLOR_BGR2GRAY)
print(gimg.shape)
plt.imshow(gimg, cmap='gray')
```

```python
blurimg = cv2.GaussianBlur(gimg, (3, 3), 1)
#outimg = cv2.resize(blurimg, None, fx = 0.5, fy = 0.5, interpolation=cv2.INTER_NEAREST)
outimg = cv2.resize(blurimg, None, fx = 0.3, fy = 0.3, interpolation=cv2.INTER_LINEAR)
#outimg = cv2.resize(blurimg, None, fx = 0.3, fy = 0.3, interpolation=cv2.INTER_CUBIC)

cv2_imshow(outimg)
```
